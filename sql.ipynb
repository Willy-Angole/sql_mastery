```python
import sqlite3
import pandas as pd
#connect to database
conn = sqlite3.connect("Jobs.db")

#read tables in the database
cursor = conn.execute('SELECT * FROM recent_grads')
```

```python
#fetch all
result = cursor.fetchall()

#convert sql query to pandas dataframe
df = pd.read_sql('SELECT * FROM recent_grads', conn)
df.head()
```

```python
df1 = pd.read_sql('SELECT Total, Major FROM recent_grads LIMIT 6', conn)
print(df1)
```

```python
#use of where clause
df2 = pd.read_sql('SELECT Major, Major_category FROM recent_grads WHERE Total < 1000', conn)
df2.head()
```

```python
df3 = pd.read_sql("SELECT Major FROM recent_grads WHERE Major_category = 'Biology & Life Science' AND ShareWomen < 0.5", conn)
df3
```

```python
df4 = pd.read_sql("SELECT Major, ShareWomen, Median FROM recent_grads WHERE Median > 10000 OR ShareWomen < 0.5 LIMIT 10", conn)
print(df4)
```

```python
df5 = pd.read_sql("SELECT Major, Major_category, Unemployed FROM recent_grads ORDER BY Unemployed DESC LIMIT 10", conn)
print(df5)
```

```python
df6 = pd.read_sql("SELECT Major, Major_category, Unemployment_rate FROM recent_grads ORDER BY Unemployment_rate DESC", conn)
print(df6)
```
